```python
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python

t = 7*10
h=0.01
N= int(t/h)
ve = np.ones(N+1)
xe = np.zeros(N+1)
T1 = [0]
#euler
for i in range(N):
    xe[i+1] = xe[i]+ ve[i]*h
    ve[i+1] = ve[i] - xe[i]*h
    T1.append(T1[i]+h)
plt.plot (T1,xe)    
plt.legend(['euler'])
plt.xlabel(" time")
plt.ylabel(" x ")
plt.show() 

plt.plot (xe,ve)    
plt.xlabel(" x ")
plt.ylabel(" v ")
plt.show()    
```

```python
#euler-cramer
v_ec = np.ones(N+1)
x_ec = np.zeros(N+1)
for i in range(N):
    v_ec[i+1] = v_ec[i] - x_ec[i]*h
    x_ec[i+1] = x_ec[i]+ v_ec[i+1]*h
    
plt.plot (T1,x_ec)    
plt.legend(['euler-cramer'])
plt.xlabel(" time")
plt.ylabel(" x ")
plt.show() 

plt.plot (x_ec,v_ec)    
plt.xlabel(" x ")
plt.ylabel(" v ")
plt.show()    
```

```python
#verlet
t=70
h=0.01
N= int(t/h)
T2 = [0]
v_v = np.ones(N+1)
x_v = np.zeros(N+1)
x_v[1] = x_ec[1]

for i in range(1,N):
    x_v[i+1] = 2*x_v[i]-x_v[i-1] - x_v[i]*h*h
    v_v[i] = (x_v[i+1]- x_v[i-1])/(2*h)
    T2.append(T2[i-1]+h)
plt.plot (T2,x_v[:-1])    
plt.legend(['verlet'])
plt.xlabel(" time")
plt.ylabel(" x ")
plt.show() 

plt.plot (x_v,v_v)    
plt.xlabel(" x ")
plt.ylabel(" v ")
plt.show()    
```

```python
#speed-verlet
h=0.001
N= int(t/h)
v_sv = np.ones(N+1)
x_sv = np.zeros(N+1)
x_sv[1] = x_v[1]
T3 = [0]

for i in range(1,N):
    x_sv[i+1] = x_sv[i]+v_sv[i-1]*h -0.5* x_sv[i]*h*h
    v_sv[i+1] = v_sv[i]-0.5*( x_sv[i+1]+x_sv[i])*h
    T3.append(T3[i-1]+h)
plt.plot (T3,x_sv[:-1])    
plt.legend(['speed verlet'])
plt.xlabel(" time")
plt.ylabel(" x ")
plt.show() 

plt.plot (x_sv,v_sv)    
plt.xlabel(" x ")
plt.ylabel(" v ")
plt.show()   
```

```python
#biman
t =70
h=0.01
N= int(t/h)
v_b = np.ones(N+1)
x_b = np.zeros(N+1)
x_b[1] = x_v[1]
T4 = [0]

for i in range(1,N):
    x_b[i+1] = x_b[i]+v_b[i]*h -(1/6)* (4*x_b[i]-x_b[i-1])*h*h
    v_b[i+1] = v_b[i]-(1/6)*(2* x_b[i+1]+5*x_b[i]-x_b[i-1])*h
    T4.append(T4[i-1]+h)
plt.plot (T4,x_b[:-1])    
plt.legend(['biman'])
plt.xlabel(" time")
plt.ylabel(" x ")
plt.show() 

plt.plot (x_b,v_b)    
plt.xlabel(" x ")
plt.ylabel(" v ")
plt.show()   
```

```python
plt.plot (T4,x_b[:-1])
plt.plot (T3,x_sv[:-1])
plt.plot (T1,xe) 
plt.plot (T1,x_ec) 
plt.plot (T2,x_v[:-1])
plt.legend(['biman','speed verlet','euler','euler kramer','verlet'])
plt.xlabel(" time")
plt.ylabel(" x ")
plt.show() 
```
