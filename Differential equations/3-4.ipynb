```python
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
#euler
v= 10 
r = 3000
c = 10**(-6)
h = 0.5*10**(-7)
start_time = datetime.now()
for t in [0.0005,0.05]:
    N = int(t/h)
    qe = np.zeros (N+1)
    T = [0]

    #t=0.5*10*(-3)
    #qf = 10**(-5)*(1-np.exp(-t/(0.003)))
    
    for  i in range (N):
        qe[i+1] = qe[i]+ (v/r - qe[i]/(c*r))*h
        T.append(T[i]+h)
    end_time = datetime.now()
    print('Duration: {}'.format(end_time - start_time))

    plt.plot (T,qe)    
    plt.legend(['t='+str (t)])
    plt.xlabel(" time")
    plt.ylabel("q ")
    plt.show()
```

```python
#euler
v= 10 
r = 3000
c = 10**(-6)
h = 10**(-5)
t = 0.02
N = int(t/h)
qe = np.zeros (N+1)
T = [0]
for  i in range (N):
    qe[i+1] = qe[i]+ (v/r - qe[i]/(c*r))*h
    T.append(T[i]+h)
```

```python
#charge

y = np.zeros(N+1)
Q =[]
y[0] = qe[0]
y[1] = qe[1]
for i in range (1,N):
    y[i+1] = y[i-1] +2*h*(v/r - y[i]/(c*r))
for i in range (N):
    Q.append(v*c*(1-np.exp(-(T[i])/0.003)))
plt.plot (T,y)
plt.plot (T[:-1],Q)
plt.legend(['napaydari','tahlili','t='+str (t)])
plt.xlabel(" time")
plt.ylabel("q- charge")
plt.show()
```

```python
#discharge

y = np.zeros(N+1)
Q =[]
y[0] = qe[N]
print(y[0])
y[1] = qe[N-1]
for i in range (1,N):
    y[i+1] = y[i-1] +2*h*( - y[i]/(c*r))
for j in range (N):
    Q.append((qe[N])*np.exp(-(T[j])/0.003))
plt.plot (T,y)
plt.plot (T[:-1],Q)
plt.legend(['napaydari','tahlili','t='+str (t)])
plt.xlabel(" time")
plt.ylabel("q-discharging")
plt.show()
print(Q)
```
